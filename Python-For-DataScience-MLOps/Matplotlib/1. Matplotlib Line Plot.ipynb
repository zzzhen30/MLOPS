### Adjusting plots

```python
#color
plt.plot(x,y ,color = "r")
plt.show()
```

```python
#linestyle
plt.plot(x,y,color = "r",linestyle="--")
plt.plot(x,y+1,color = "blue",linestyle="-.")
plt.plot(x,y+2,color = "black",linestyle=":")
plt.show()
```

```python
#marker
plt.plot(x,y,color = "r",linestyle="--",marker="+")
plt.show()
```

```python
plt.plot(x,y, linestyle = '-.' , color = 'g')
plt.plot(x,z, linestyle = '--' , color = 'r')
plt.show()
```

```python
#legend
plt.plot(x,y, label = "sine function" )
plt.plot(x,z , label = "cosine function")
plt.legend()
plt.show()
```

```python
#Format [marker][color][linestyle]
plt.plot(x,y,'og--')
plt.show()
```

```python
plt.plot(x,y,'^k:')
plt.show()
```

```python
#Title
plt.plot(x,y,'-b', label = "Sine Function")
plt.title("Sine Function heading")
plt.legend()
plt.show()
```

```python
#Axes limits
plt.plot(x,y,'-b', label = "Sine Function")
plt.title("Sine Function heading")
plt.xlabel("X axis label")
plt.ylabel("Y axis label")
plt.legend()
plt.show()
```

```python
#Axes limits
plt.plot(x,y,'-b', label = "Sine Function")
plt.title("Sine Function heading")
plt.xlim(-10,10)
plt.ylim(-5,5)
plt.legend()
plt.show()
```
