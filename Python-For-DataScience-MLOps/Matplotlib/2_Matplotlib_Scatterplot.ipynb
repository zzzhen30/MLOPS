# Setting up the Notebook

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

## Scatter Plot - `plt.plot()`

```python
x = np.linspace(0,10,25)
y = np.cos(x)
plt.plot(x, y, 'o', color='black') #marker sytle
plt.show()
```

```python
np.random.rand(5)
```

```python
marker_styles = ['o', 'x', '+', 'v', '^', 's', 'd','p','*']

for marker in marker_styles:
    plt.plot(np.random.rand(5), np.random.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend()
plt.xlim(0, 2)
plt.show()
```

```python
plt.plot(x, np.sin(x), 'p', markersize=20,
         markerfacecolor='magenta',
         markeredgecolor='black',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2)
plt.show()
```

#  ``plt.scatter``

```python
plt.scatter(x, y, marker='o')
plt.show()
```

```python
x = np.linspace(0,10,25)
y = np.sin(x)
plt.scatter(x, y,marker = "D",cmap = "Set1", c=x)
#plt.colorbar()
plt.show()
```

https://matplotlib.org/tutorials/colors/colormaps.html

```python
x = np.linspace(0,10,25)
y = np.sin(x)
plt.scatter(x, y,marker = "D",cmap = "viridis", c=x)
plt.colorbar()
plt.show()
```

```python
x = np.linspace(0,10,25)
y = np.sin(x)
plt.scatter(x, y,marker = "D",cmap = "cool", c=y)
plt.colorbar()
plt.show()
```
