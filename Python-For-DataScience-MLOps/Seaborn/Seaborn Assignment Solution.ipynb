```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
iris = sns.load_dataset("iris")
```

```python
iris
```

1. Plot the distribution of sepal_length in the given dataset. (distplot)

```python
sns.distplot(iris["sepal_length"])
plt.show()
```

2. Plot the distribution of petal_length in the given dataset. (distplot)

```python
sns.distplot(iris["petal_length"])
plt.show()
```

3. Make a joint plot between Sepal Length & Sepal Width.

```python
sns.jointplot(iris["sepal_length"],iris["sepal_width"])
plt.show()
```

4. Give additional attribute kind = "hex" in the above plot.

```python
sns.jointplot(iris["sepal_length"],iris["sepal_width"],kind="hex")
plt.show()
```

5. Make pairplot for the dataframe

```python
sns.pairplot(iris)
plt.show()
```

6. Make striplot between species and sepal_length.

```python
sns.stripplot(iris['species'], iris['sepal_length'])
plt.show()
```

7. Make bar plot with species and sepal_width

```python
sns.barplot(iris['species'], iris['sepal_width'])
plt.show()
```

8. Create Box plot with: Species and Sepal length

```python
sns.boxplot(iris["species"],iris["sepal_length"])
plt.show()
```
