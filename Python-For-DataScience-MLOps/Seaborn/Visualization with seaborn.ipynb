## Distributions - Univariate

seaborn.distplot() --> draw a histogram and fit a kernel density estimate (KDE).  



```python
sns.distplot(df['city_mpg'])
plt.show()
```

```python
sns.distplot(df['city_mpg'],kde= False)
plt.show()
```

```python
#kde off and tic mark on x-axis
sns.distplot(df['city_mpg'], kde=False, rug=True)
plt.show()
```

## Distributions - Bivariate

Visualize a relationship between two variables. 

sns.jointplot() --> scatterplot with the histograms of each next to the appropriate axes.



```python
sns.jointplot(df['engine_size'], df['horsepower'])
plt.show()
```

```python
sns.jointplot(df['highway_mpg'], df['city_mpg'])
plt.show()
#does it make sense ?
```

### Plots - Hex Bin 

Splits the 2D area into hexagons and color is determined based on no. of points

```python
sns.jointplot(df['engine_size'], df['horsepower'], kind="hex")
plt.show()
```

### KDE with jointplot


```python
sns.jointplot(df['engine_size'], df['horsepower'], kind="kde")
plt.show()
```

## Visualizing pairwise relationships in a dataset

To plot multiple pairwise scatterplots in a dataset, you can use the pairplot() function. This creates a matrix of axes and shows the relationship for each pair of columns in a DataFrame, it also draws the histogram of each variable on the diagonal Axes:

Be careful about toggle scrolling in the cell menu!!!

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df[['highway_mpg', 'engine_size', 'horsepower']])
plt.show()
```

## categorical data - stripplot()



```python
df['fuel_type'].value_counts()
```

```python
sns.stripplot(df['fuel_type'], df['horsepower'],jitter=False)
plt.show()
```

## Boxplots - sns.boxplot()

```python
sns.boxplot( df['city_mpg'])
plt.show()
```

```python
#Used for identifying the outliers in the data
sns.boxplot(df['number_of_doors'], df['city_mpg'])
plt.show()
```

```python
sns.boxplot(df['number_of_doors'], df['highway_mpg'], hue=df['fuel_type'])
plt.show()
```

## Bar plots --> sns.barplot()

```python
#estimate of central tendency for a numeric variable
sns.barplot(df['body_style'], df['highway_mpg'], hue=df['fuel_type'])
plt.show()
```

```python
sns.barplot(df['highway_mpg'],df['body_style'], hue=df['fuel_type'])
plt.show()
```

countplot() --> Count of observations in each category

```python
sns.countplot(df['number_of_doors'],hue=df['fuel_type'])
plt.show()
```

```python
sns.countplot(df['fuel_type'],hue=df['body_style']);
```

## Point plots

This plots the point estimate and confidence interval. 
Along with this, the pointplot connects points from the same hue category. This makes it easy to see how the main relationship is changing as a function of a second variable, because your eyes are quite good at picking up on differences of slopes:

```python
sns.pointplot(df['body_style'], df['highway_mpg'], hue=df['number_of_doors'])
plt.show()
```

```python
sns.pointplot(df['body_style'], df['highway_mpg'], hue=df['number_of_doors'],ci = False)
plt.show()
```

## Drawing multi-panel categorical plots

```python
df.info()
```

```python
sns.catplot(x="fuel_type",
               y = "horsepower",
               hue="number_of_doors", 
               col="drive_wheels", 
               data=df, 
               kind="box")
plt.show()
```

```python
sns.catplot(x="fuel_type",
               y = "horsepower",
               hue="number_of_doors", 
               col="drive_wheels", 
               data=df, 
               kind="point")
plt.show()
```

Various types of kind input : {``point``, ``bar``, ``count``, ``box``, ``violin``, ``strip``}

## Function to draw linear regression models

lmplot() is one of the most widely used function to quickly plot the Linear Relationship between 2 variables

```python
sns.lmplot(y="horsepower", x="engine_size", data=df)
plt.show()
```

```python
sns.lmplot(y="horsepower", x="price", data=df)
plt.show()
```

```python
sns.lmplot(y="horsepower", x="price",hue="fuel_type", data=df);
```
